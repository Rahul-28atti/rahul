```python
rooms = [2, 3, 4, 3, 5]
trees = [1, 2, 3, 2, 4]
print("Rooms:", rooms)
print("Trees:", trees)

house_types = ("Apartment", "Villa", "Bungalow", "Apartment", "Villa")

# Tuple: imm
print("House types:", house_types)


# Dictionary combining all features
house_data = {"Rooms": rooms, "Trees": trees, "Type": house_types}
print("House data dictionary:", house_data)
```

```python
import pandas as pd
df = pd.DataFrame(house_data)
print(df)

print(df.describe())

print("Rooms column:", df['Rooms'])

df['Total_Features'] = df['Rooms'] * df['Trees']
print(df)
```

```python
import matplotlib.pyplot as plt # Import Matplotlib for plotting

plt.figure(figsize=(8,4))
plt.plot(df ['Rooms'], label="Rooms", marker='o') # Plot rooms
plt.plot(df['Trees'], label="Trees", marker='s') # Plot trees
plt.title("Number of Rooms and Trees")
plt.xlabel("House Index")
plt.ylabel("Count")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.bar(df.index, df ['Total_Features'], color='pink')

plt.title("Total Features per House")

plt.xlabel("House Index")
plt.ylabel("Total Features")

plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.scatter(df ['Rooms'], df ['Trees'], color='red', s=100) # Scatter: 

plt.title("Scatter Plot: Rooms vs Trees")

plt.xlabel("Rooms")

plt.ylabel("Trees")
plt.grid(True)

plt.show()
```
